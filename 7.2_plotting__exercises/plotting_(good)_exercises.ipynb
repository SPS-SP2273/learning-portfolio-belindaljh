```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# original code given --> CHANGE STRINGS TO RAW STRING: r'$mathcode$'

#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, num=100, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2,
                        figsize=(12, 8),  sharey='row')

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label=r'$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='r',
                linestyle=':', label=r'$\cos^2 x$')
axes[0, 0].set_title(r'$\cos x$ & $\cos^2x$')
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].fill_between(x, cos_x, -1, color='g', alpha=.125)
axes[0, 0].set_xlabel('Angle (radians)')
axes[0, 0].text(0, 0, 'I am [0, 0]!', fontsize=30,
                horizontalalignment='center')

#------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='g', label=r'$\sin x$')
axes[0, 1].fill_between(x, cos_x, -2, color='r', alpha=.125)
axes[0, 1].plot(x, sin_x**2, label=r'$\sin^2 x$')
axes[0, 1].set_ylabel('Cosine Value')
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].legend(loc='lower right', frameon=False)
axes[0, 1].text(0, 0, 'I am [0, 1]!', fontsize=30,
                horizontalalignment='center')

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label=r'$e^{-x}\cos 5x$')
axes[1, 0].fill_between(x, fun1_x, 0, color='b', alpha=.125)
axes[1, 0].set_title(r'$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Cosine Value')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels([r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
axes[1, 0].legend()
axes[1, 0].text(0, 0, 'I am [1, 0]!', fontsize=30,
                horizontalalignment='center')

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color='y', label=r'$e^{-x}\sin 2x$')
axes[1, 1].set_title(r'$e^{-x}\sin 2x$')
axes[1, 1].fill_between(x, fun2_x, 10, color='y', alpha=.125)
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].legend()
axes[1, 1].text(0, 0, 'I am [1, 1]!', fontsize=30,
                horizontalalignment='center')

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
    a.grid(alpha=.5)

# plt.tight_layout()
plt.show(block=False)
```

```python
# debugging 

#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, 100, endpoint=True)      # endpoint=True ensures that the last value in the array, 
                                                        # i.e. np.pi, is included in the sequence. 
cos_x = np.cos(x)                                       # use numpy package to do the math and define the variables
sin_x = np.sin(x)                                           # eg. np.cos(x) computes cosine of each value in array x, then whole new array 
fun1_x = np.exp(-x) * np.cos(5 * x)                         # gets stored in variable cos_x
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2,                              # set no. of rows&columns: 2 subplots in each row and in each column
                        figsize=(13, 8),  sharey='row', sharex='col')   # sharey / sharex = allow subplots to share the same x and y axis, 
                                                                        # either across columns or rows

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label=r'$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='r',
                linestyle=':', label=r'$\cos^2 x$')
axes[0, 0].set_title(r'$\cos x$ & $\cos^2x$')
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].fill_between(x, cos_x, -2, color='r', alpha=.125)
axes[0, 0].set_xlim(-np.pi, np.pi)                          # Matplotlib automatically pads axis limits slightly beyond values in array,
axes[0, 0].legend(loc='upper right')                        # hence there'll be a small gap. Use set_xlim to make x-axis limit match values exactly

#------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='g', label=r'$\sin x$')
axes[0, 1].fill_between(x, sin_x, -2, color='g', alpha=.125)    #fill_between(first value in array x, area under which graph, until where along y-axis, ...)
axes[0, 1].set_title(r'$\sin x$ & $\sin^2x$')
axes[0, 1].plot(x, sin_x**2, label=r'$\sin^2 x$')
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].set_xlim(-np.pi, np.pi)
axes[0, 1].legend(loc='upper right')

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label=r'$e^{-x}\cos 5x$')
axes[1, 0].fill_between(x, fun1_x, -25, color='b', alpha=.125)
axes[1, 0].set_title(r'$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels([r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
axes[1, 0].set_ylabel('Cosine Value')
axes[1, 0].set_xlim(-np.pi, np.pi)
axes[1, 0].legend()

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color='y', label=r'$e^{-x}\sin 2x$')
axes[1, 1].set_title(r'$e^{-x}\sin 2x$')
axes[1, 1].fill_between(x, fun2_x, -25, color='y', alpha=.125)
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].set_xticklabels([r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
axes[1, 1].set_xlim(-np.pi, np.pi)
axes[1, 1].legend()
axes[1, 1].set_xlabel('Angle (radians)')

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
    a.grid(alpha=.5)

# plt.tight_layout() + saving figure as jpg using plt.savefig()
plt.savefig('Plotting(good) Exercise 1 Final Subplots.jpg',dpi=150)         #must be called before plt.show(), otherwise will
                                                                            #only save a blank, white image. 
plt.show(block=False)
plt.tight_layout()
```
