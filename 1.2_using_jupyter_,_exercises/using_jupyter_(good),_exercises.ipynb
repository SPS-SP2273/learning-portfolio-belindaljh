```python
# first code example
squares_1 = [i**2 for i in range(10)]

# second code example
squares_2 = list()
for i in range(10):
    squares_2.append(i**2)

# verify that both give same results
print(squares_1)
print(squares_2)
```

```python
# measure the execution time for first code example
%%timeit
squares_1 = [i**2 for i in range(10)]

# 818ns --> average time taken for one loop of the code 
# 66.3ns --> variation of each run, i.e. how much the timing fluctuated 
# 1000000 loops --> the number of times/loops the code was ran (1 run)
# 7 runs --> repeat above for 7 times 
```
